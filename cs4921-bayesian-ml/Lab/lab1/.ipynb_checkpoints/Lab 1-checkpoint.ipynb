<span>2.</span> Use the derivative of $a^2$ and an initial a of 1 to find the $\underset{x}{\operatorname{argmin}}a^2$ using Pytorch/Tensorflow and gradient
descent.

```python
x = tf.Variable(1.0)
thresh = 0.00001
optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)

with tf.GradientTape() as tape:
    loss = tf.square(x)
    gradients = tape.gradient(loss, [x])

while abs(gradients[0].numpy()) > thresh:
    with tf.GradientTape() as tape:
        loss = tf.square(x)
        gradients = tape.gradient(loss, [x])
    optimizer.apply_gradients(zip(gradients, [x]))

print('Output: Minimum at x =', x.numpy())
```

> Stochastic Gradient Descent (SGD) found an approximate minimum at x = 0.000004819498.

<span>4.</span> Use the derivative of $sin(x) − sin(0.5x)$ and an initial $x$ of 0.5 to try and find $\underset{x}{\operatorname{argmin}}sin(x) − sin(0.5x)$ using Pytorch/Tensorflow and gradient
descent.

```python
x = tf.Variable(0.5)
thresh = 0.0001
optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)

with tf.GradientTape() as tape:
    loss = tf.sin(x) - tf.sin(0.5*x)
    gradients = tape.gradient(loss, [x])

while abs(gradients[0].numpy()) > thresh:
    with tf.GradientTape() as tape:
        loss = tf.sin(x) - tf.sin(0.5*x)
        gradients = tape.gradient(loss, [x])
    optimizer.apply_gradients(zip(gradients, [x]))

print('Output: Minimum at x =', x.numpy())
```

```python
def function(x):
    x=x.numpy()
    return sin(x)-sin(0.5*x)
print('Value of function at x:',function(x))
```

Implement linear layer to input x, use GradientTape

<span>7.</span> Given $g(w_1, w_2, w_3, w_4, x_1, x_2, x_3) = w_1x_1+w_2x_2+w_3x_3+w_4$, an initial $x$ of (1,2,3), and a randomly intialized $w$, find $\frac{\partial g^2}{\partial w_1},\frac{\partial g^2}{\partial w_2},\frac{\partial g^2}{\partial w_3}$ using Pytorch/Tensorflow.

```python
x = tf.constant([[1., 2., 3.]])

layer = tf.keras.layers.Dense(1, activation=None)
with tf.GradientTape(persistent=True) as tape:
    g = layer(x)
    g2 = tf.square(g)
    
gradients = tape.gradient(g2, layer.trainable_variables)
print(gradients[0].numpy().T)
```

<span>b.</span> Use Pytorch, the Gaussian negative log-likelihood, and gradient descent to compute the mean and
the standard deviation of the aircraft costs.

```python
x = costs_array
print(x)
epochs = 100
sigma = tf.Variable(np.random.uniform())
mu = tf.Variable(np.random.uniform())

thresh = 0.0001
optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)

for epoch in range(epochs):
    with tf.GradientTape() as tape:
        loss = tf.math.log(1/(tf.math.sqrt(2*tf.constant(pi)*sigma)))+tf.square(tf.math.abs(x-mu))/(2*tf.square(sigma))
        loss_sum = -tf.reduce_sum(loss)
        gradients = tape.gradient(loss, [sigma, mu])
        optimizer.apply_gradients(zip(gradients, [sigma, mu]))

print('sigma:', sigma.numpy(), 'mu:', mu.numpy())
print('gradients:', gradients)
```
